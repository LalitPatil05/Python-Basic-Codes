## Types of Function argument in Python :
1. Required arguments
2. Default arguments
3. Keyword arguments
4. Arbitrary arguments

````python
# 1. Required Arguments

def intro(instructor, course_name):
    print(f"Welcome to, {course_name} by {instructor}")

intro("Lalit", "Python Tutorial")
````

````python
# 2. Default Arguments 
# you can assign a default values to arguments in function definition.
def greetings(name="User"):
    print("Hello,",name+"!")

greetings()
greetings("Lalit")
````

````python
# 3. Keyword arguments 
# when calling a function, you can specify arguments by the parameter name.
# it can be given in any order.

def add(a,b):
    print(a/b)

add(b=20,a=10) # keyword argument
add(10,20) # positional argument
````

````python
# 4. Arbitrary Arguments
# if you are unsure how many arguments will be passed, use *args to accept any number of positional arguments.
# arguments are stored as a tuple.

def add(*args):
    print(sum(args))

add(1,2,3)
add(1,2,3,4,5)
````

````python
''' Set in Python
Set is unique items in python. set do not allow duplicate items and do not mainitain any particular order so it can't be indexed.

Characterstics of Set :-

1. Unordered
2. Unique Elements
3. Mutable
4. Immutable Elements

Example :- vowels = {'a', 'e', 'i', 'o', 'u'}
'''
````

````python
# Creating a Set

vowels = {'a', 'e', 'i', 'o', 'u'}
print(vowels)
print(type(vowels))

set1 = set([1,2,3,4,5])
print(set1)
print(type(set1))

emptySet = set()
print(emptySet)
print(type(emptySet))
````

````python
# Set Operation

# 1. Adding Elements

numset = {10,20,30,40,50}
print(numset)
print(type(numset))

numset.add(60)
print(numset)

# 2. Removing Elements

numset.remove(60)
print(numset)

# 3. Discard Element

numset.discard(20)
print(numset)
````

````python
# Set Methods

# 1. Union :- combine elements from two sets, removing duplicates

set_a = {1,2,3}
set_b = {3,4,5}
union_set = set_a.union(set_b)
print(union_set)
print(set_a | set_b) # alternate way to union of two sets
````

````python
# 2. Intersection Method :- Includes only elements present in both sets
intersection_set = set_a.intersection(set_b)
print(intersection_set)
print(set_a & set_b) # alternate way
print(type(intersection_set))
````

````python
# 3. Difference Set :- Elements present in the first set but not in the second.

differenceSet = set_a.difference(set_b)
print(differenceSet)
print(type(differenceSet))
print(set_a - set_b) # alternate way

differenceSet2 = set_b.difference(set_a)
print(differenceSet2)
print(set_b - set_a) # alternate way
````

````python
# Symmetric differenece :- Elements in either set, but not in both.

sym_diff_set = set_a.symmetric_difference(set_b)
print(sym_diff_set)
print(type(sym_diff_set))
print(set_a ^ set_b)
print(set_b ^ set_a)
````

````python
# Practice Set

# Q.1 find the intersection (common elements) of two lists
def intersection_loop(lst1, lst2):
    common_list = []
    for item in lst1:
        if item in lst2 and item not in common_list:
            common_list.append(item)
    return common_list

# function calling

list1 = [1,2,3,4]
list2 = [1,3,5,7]
print(intersection_loop(list1,list2))
````

````python
# Using list comprehension 

def intersection_loop_comp(lst1, lst2):
    return [item for item in lst1 if item in lst2]

print(intersection_loop(list1,list2))
print(intersection_loop_comp(list1,list2))
````

````python
# 2. return or find the most frequent element arised in the list

def most_freq(lst):
    max_count = 0
    most_freq = None
    for item in lst:
        count = lst.count(item)
        if count > max_count:
            max_count = count
            most_freq = item
    return most_freq

list1 = [1,2,2,3,3,3,4]
print(most_freq(list1))
````

````python
# find the cumulative sum of a list

numbers = [1,2,3,4,5]

def cumulative_sum(lst):
    cum_sum = []
    total = 0
    for num in lst:
        total = total + num
        cum_sum.append(total)
    return cum_sum

print(cumulative_sum(numbers))
````

````python
# remove duplicates into the list

num = [1,2,2,3,4,5,6,3,2,1]

def remove_duplicates(lst):
    unique = []
    seen = set()
    for item in lst:
        if item not in seen:
            unique.append(item)
            seen.add(item)
    return unique

print(remove_duplicates(num))
````

````python
# find the most frequent value into the dictionary

data = {'a':1, 'b':2, 'c':1, 'd':3, 'e':1}

def most_time(dct):
    frequency = {}
    for value in dct.values():
        if value not in frequency:
            frequency[value] = 0
        frequency[value] += 1
        max_value = max(frequency, key=frequency.get)
        return max_value

print(most_time(data))
````

````python
# Q.7 Merge Dictionaries with Summation 

dict1 = {'a':10, 'b':20, 'c':30}
dict2 = {'b':15, 'c':35, 'd':25}

def merge_dict(dct1, dct2):
     result = dct1.copy()
     for key, value in dict2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value
     return result

print(merge_dict(dict1, dict2))
````

````python
# Q. flatten a nested dictionary 

data = {'a':{'b':{'c':42}, 'd':7}, 'e':10}
# o/p :- {a.b.c:42, a.d:7, e:10}

def flatten_dict(data, parent_key = '', sep='.'):
    items = {}
    for key, value in data.items():
        new_key = f"{parent_key}{sep}{key}" if parent_key else key
        if isinstance(value, dict):
            items.update(flatten_dict(value, new_key, sep))
        else:
            items[new_key] = value
    return items

print(flatten_dict(data))
````

````python
# Q.9 Sort a Dictionary by values in ascending order

data = {'a':5, 'b':9, 'c':2, 'd':7}

def sort_by_values(data):
    sorted_items = sorted(data.items(), key = lambda item : item[1])
    return {key:value for key, value in sorted_items}

# function calling

print(sort_by_values(data))
````

````python
# Q.9 Sort a Dictionary by values in descending order

data = {'a':5, 'b':9, 'c':2, 'd':7}

def sort_by_values(data):
    sorted_items = sorted(data.items(), key = lambda item : item[1], reverse=True)
    return {key:value for key, value in sorted_items}

# function calling

print(sort_by_values(data))
````

````python
class Student:
    def __init__(self,name, grade, percentage, team):
        self.name = name
        self.grade = grade
        self.percentage = percentage
        self.team = team

    def student_details(self):
        print(f"{self.name} is in {self.grade} grade with {self.percentage}%, from team {self.team}")

# Object Creation

team1 = "A"
team2 = "B"

student1 = Student("Puja", "A+", 97, team1)
student1.student_details()

student2 = Student("Lalit", "B", 77, team2)
student2.student_details()
````

````python
# Abstraction in Python 
# Hiding unnecessary details from users through class, methods.

class Student:
    def __init__(self,name, grade, percentage, team):
        self.name = name
        self.grade = grade
        self.percentage = percentage
        self.team = team

    def student_details(self): # method - abstraction
        print(f"{self.name} is in {self.grade} grade with {self.percentage+2}%, from team {self.team}")

# Object Creation

team1 = "A"
team2 = "B"

student1 = Student("Puja", "A+", 97, team1)
student1.student_details()

student2 = Student("Lalit", "B", 77, team2)
student2.student_details()
````

````python
# Encapsulation in Python 
# Restrict access to certain attributes or methods to protect data and enforce controlled access

class Student:
    def __init__(self,name, grade, percentage, team):
        self.name = name  # attributes
        self.grade = grade
        self.__percentage = percentage # double underscore limits access
        self.team = team

    def get_percentage(self):
        return self.__percentage

    def student_details(self): # method - abstraction
        print(f"{self.name} is in {self.grade} grade with {self.percentage}%, from team {self.team}")

# Object Creation

team1 = "A"
team2 = "B"

student1 = Student("Puja", "A+", 97, team1)
print(student1.get_percentage())
# print(student1.percentage) Error
````

````python
# Inheritance in Python 
# allows one class (child) to reuse the prop and methods of another class (parent).

# Parent Class 

class Student:
    def __init__(self,name, grade, percentage, team):
        self.name = name  # attribute
        self.grade = grade
        self.percentage = percentage # double underscore limits access
        self.team = team

    def student_details(self): # method - abstraction
        print(f"{self.name} has {self.grade} grade with {self.percentage}%, from team {self.team}")

# Child Class

# GraduateStudent child class inherit properties and methods from Student Parent Class.
class GraduateStudent(Student):
    def __init__(self, name, grade, percentage, team, stream):
        # Parameters from parent class and new parameters in child class
        super().__init__(name, grade, percentage, team) # class parent class __init__
        self.stream = stream # new attribute in child class

    def student_details(self): # method
        super().student_details() # method inherit from parent class
        print(f"Graduated in Stream {self.stream}")
        
# Object Creation

team1 = "Vikas"
stream = "ComputerSciennce"

graduateStudent1 = GraduateStudent('Sumit', "A", 95, team1, stream)
graduateStudent1.student_details()

student1 = Student("Kunal", "A+", 88, "KK")
student1.student_details()
````

````python
class Calculator:
    def multiply(self, a=1, b=1, *args):
        result = a * b
        for num in args:
            result *= num
        return result

# Create object
calc = Calculator()

# Using default arguments
print(calc.multiply())            
print(calc.multiply(4))           

# Using multiple arguments
print(calc.multiply(2, 3))       
print(calc.multiply(2, 3, 4))
````
